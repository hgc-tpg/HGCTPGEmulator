```python
def getTCinfo(idx,modHash):
    theTC = all_tcs[(all_tcs.tcaddress==idx)&(all_tcs.module_hash==modHash)]
    x = theTC['x'].values[0]
    y = theTC['y'].values[0]
    z = theTC['z'].values[0]
    return x,y,z
```

```python
for modHash in mod_stimuli:
    for event_tcs in mod_stimuli[modHash]:
        for idx, e in enumerate(event_tcs):
            x,y,z = getTCinfo(idx,modHash)
            print("l1thgc::HGCalTriggerCell_SA(x={0},y={1},z={2},e={3})".format(x,y,z,e))
```
